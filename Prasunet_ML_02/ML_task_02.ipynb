```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Define a color palette for clusters
palette = sns.color_palette("husl", n_colors=len(df['cluster'].unique()))

# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x="Annual Income (k$)", y="Spending Score (1-100)", data=df, hue="cluster",
    palette=palette, s=60, alpha=0.8, edgecolor='w', linewidth=0.5
)

# Plot cluster centroids
plt.scatter(
    x=kmeans.cluster_centers_[:, 0],
    y=kmeans.cluster_centers_[:, 1],
    color="red",
    s=150,  # Size of centroids
    marker="X",  # Marker style for centroids
    edgecolor='k',  # Edge color for centroids
    label="Centroids"
)

# Annotate centroids with their cluster number
for i, (x, y) in enumerate(kmeans.cluster_centers_):
    plt.text(
        x, y, f'Cluster {i+1}', color='black', fontsize=12,
        ha='center', va='center', fontweight='bold'
    )

# Add titles, labels, and gridlines
plt.title("K-Means Clustering - Clusters of Customers", fontsize=16)
plt.xlabel("Annual Income (k$)", fontsize=14)
plt.ylabel("Spending Score (1-100)", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
